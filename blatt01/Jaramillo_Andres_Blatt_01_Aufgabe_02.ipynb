```python
import math

f1 = lambda y, z: math.cos(math.pi/13 * y) + math.cos(3*math.pi/13 * y) + math.cos(4*math.pi/13 * y) - z
f2 = lambda y, z: 8*z**2 + 4*z - 3*y

df1_dy = lambda y, z: -(math.pi/13)*math.sin(math.pi/13 * y) - (3*math.pi/13)*math.sin(3*math.pi/13 * y) - (4*math.pi/13)*math.sin(4*math.pi/13 * y)
df1_dz = lambda y, z: -1.0

df2_dy = lambda y, z: -3.0
df2_dz = lambda y, z: 16*z + 4.0

Df = lambda y,z: [[df1_dy(y,z), df1_dz(y,z)],
                  [df2_dy(y,z), df2_dz(y,z)]];

f = lambda y, z: [f1(y,z), f2(y,z)]

y, z = [2, 1]
print(f"f({y},{z}) = {f(y,z)}, Df({y},{z}) = {Df(y, z)}")
```

```python
import numpy as np


Df_x1 = np.array(Df(y,z))
Df_x1_inv = np.linalg.inv(Df_x1)
f_x1 = np.array(f(y,z))

x2 = Df_x1_inv @ (-1 * f_x1)

Df_x2 = np.array(Df(x2[0], x2[1]))
Df_x2_inv = np.linalg.inv(Df_x2)
f_x2 = np.array(f(x2[0], x2[1]))

x3 = Df_x2_inv @ (-1 * f_x2)

print(f"f_x2 = {f_x2}")
print(f"Df_x2 = {Df_x2}")
print(f"Df_x2_inv = {Df_x2_inv}")
print(f"x2: {x2}")
print(f"x3: {x3}")

```
