# Aufgabe 1.b)

```python
def fixpunktiteration(f, n, x_0):
    x_i = x_0
    for i in range(n):
        fx_i = f(x_i)
        print(f"x_{i + 1} = f({x_i}) = {fx_i}")
        x_i = fx_i
    return x_i
```

```python
f1 = lambda x: 5 + x - x**2
f2 = lambda x: 5/x
f3 = lambda x: 1 + x - (x**2)/5
f4 = lambda x: 0.5 * (x + 5/x)

all_f = [f1, f2, f3, f4]
for i, f in enumerate(all_f):
    print(f"now doing f_{i+1}...")
    fixpunktiteration(f, 6, 2.5)
```
